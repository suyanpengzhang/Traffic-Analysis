```python
import pickle
import pandas as pd
import warnings
import numpy as np
```

```python
with open("Arterial_ID_zip.p", "rb") as file:
    ID_zip = pickle.load(file)
with open("sensor_zip.pkl", "rb") as file:
    ID_zip_2022 = pickle.load(file)
```

```python
#Create a dataframe 
```

```python
months = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']
years = ['2020', '2022']
zips = []
new = {}
data ={}
data['Zips'] = []
for month in months:
    for year in years:
        data[f'{year}_{month}'] = []
        file_name = f'monthly_data/monthly_{year}_{month}.pkl'
        with open(file_name, 'rb') as file:
            flow = pickle.load(file)
        for zip_id in list(flow.keys()):
            if zip_id !=0:
                if zip_id[0:5] not in zips:
                    zips.append(zip_id[0:5])
                if (f'{year}_{month}',zip_id[0:5]) not in new:
                    new[(f'{year}_{month}',zip_id[0:5])] = flow[zip_id]
                else:
                    new[(f'{year}_{month}',zip_id[0:5])] = new[(f'{year}_{month}',zip_id[0:5])] + flow[zip_id]
                    #print(new[(f'{year}_{month}',zip_id[0:5])])
```

```python
for zip_ in zips:
    data['Zips'].append(zip_)
    for month in months:
        for year in years:
            if (f'{year}_{month}',zip_) not in new:
                data[f'{year}_{month}'].append(0)
            else:
                data[f'{year}_{month}'].append(np.mean(new[(f'{year}_{month}',zip_)]))
    
    
```

```python
df = pd.DataFrame(data)
  
df
file_path = f'monthly_data/regression_data.pkl'
    
# Save the dictionary for the current month as a pickle file
with open(file_path, 'wb') as file:
    pickle.dump(df, file)
```

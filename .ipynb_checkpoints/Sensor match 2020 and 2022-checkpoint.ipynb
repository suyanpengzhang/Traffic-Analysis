```python
months = ['04', '05', '06', '10', '11', '12']
years = ['2020', '2022']
zips = []
new = {}
data ={}
data['Zips'] = []
data['sensor'] = []
for month in months:
    for year in years:
        data[f'{year}_{month}'] = []
        file_name = f'monthly_data/monthly_{year}_{month}.pkl'
        with open(file_name, 'rb') as file:
            flow = pickle.load(file)
        for zip_id in list(flow.keys()):
            if zip_id !=0:
                if zip_id[0:5] not in zips:
                    zips.append(zip_id[0:5])
                if (f'{year}_{month}',zip_id[0:5]) not in new:
                    new[(f'{year}_{month}',zip_id[0:5])] = flow[zip_id]
                else:
                    new[(f'{year}_{month}',zip_id[0:5])] = new[(f'{year}_{month}',zip_id[0:5])] + flow[zip_id]
                    #print(new[(f'{year}_{month}',zip_id[0:5])])
```
